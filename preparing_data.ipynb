```python
messages = "<|im_start|>"
for line in file:
    pattern = r'\] (.+?): (.+)$'
    match = re.search(pattern, line)
    if not match:
        messages = messages + """/n{message}""".format(message=line)
    else:
        messages = messages + "<|im_end|>"
        if match[1] == "s.":
            messages = messages + """<|im_start|>assistant<|im_sep|>{message}""".format(message=match[2])
        else: 
            messages = messages + """<|im_start|>user<|im_sep|>{message}""".format(message=match[2])
    
```

```python
file2 = open("text_for_training.txt", "w")
file2.write(messages)
```
